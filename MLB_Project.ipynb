## Introduction

Since the inception of Major League Baseball (MLB) in 1871, much has changed over time. One factor, however, has remained consistent throughout the existence of the league: Those participating are typically between the ages of 20 and 42. Across this range, player performance varies. It is this relationship that this paper aims to explore. Utilizing a curated dataset from Fangraphs and Baseball Reference, I wish to examine the effects of age and aging on player performance in Major League Baseball. In exploring this relationship, I wish to gain insight into a variety of questions. The primary question, however, is as follows: __At what age—or age range—does player performance peak? Is this rule consistent across all players, or does the age at which peak performance occurs vary for the individual?__ As to the relationship between age and player performance, I expect to observe a nonlinear relationship. More specifically, I expect overall performance to steadily improve at the outset of one’s career. Once one reaches a particular age, however, such improvement will be followed by a steep decline in overall performance. Identifying the age at which this peak occurs is my primary goal. 

In addition to identifying the age at which overall performance peaks, I intend to examine player performance across several metrics. More specifically, I will examine various skills and statistics and determine their relationship to age. The primary metrics I will consider are On Base Percentage (OBP) and Stolen Bases (SB). My interest in these metrics stem from the skills they measure. OBP is a strong measure of discipline and SB is a reliable measure of speed. In analyzing these metrics, I wish to determine the relationship between age and the skills they measure. __Across alternate performance measures and skills, does the age at which players peak differ?__ 

While my over-arching analysis will be of the relationship between players’ age and on-field production, I plan to further explore the dataset. Through my exploration, I may perhaps also gain insight on additional properties related to age, such as if players’ peaks have remained consistent over time. That is, if players are currently peaking at the age of 27, does the same apply for players in 1910? Additionally, although my analysis will largely focus on hitters, it would be of interest to compare these findings to pitchers. Do pitchers and hitters experience alternate peak ages?

As I did in my exploratory data analysis, I will consider a number of performance evaluators, such as Wins Above Replacement (WAR), On Base Percentage (OBP), and Stolen Bases (SB). By examining each of these statistics individually, I can assess the impact of age across a variety of measures and skills. WAR measures overall player performance. Thus, I expect its relationship with age to be quadratic. More specifically, I expect player performance to increase in one's early 20s, peak in their late 20s, and decline sharply thereafter. OBP is a strong measure of plate discipline, a trait often attributed to veterans, or more experienced players. Thus, I expect its relationship with age to be positive and roughly linear, indicating that one's discipline increases with age. Finally, for SB, a measure of speed, I expect to observe a negative and roughly linear relationship. Young and athletic players will likely outpace their older counterparts. As I am assessing the effect of age _on_ these variables, it will act as my input variable. In three separate regression models, I will assess age's significance in predicting WAR, OBP, and SB. Age, however, is not the only predictor of player performance. Thus, I will also include measures of career performance for each player. 
### Analysis #2: Hypothesis Test and Confidence Interval
To assess the significance of age on OBP and SB, I will conduct two permutation tests. For each test, my null hypothesis is that there is no relationship between age (x) and OBP/SB (y). To simulate this, I will shuffle the OBP/SB values across the dataset. Thereafter, I will produce 100 iterations of models on the permuted data. Finally, I will compute the fraction of permuted slopes more extreme than the observed slopes. In other words, I will evaluate the likelihood of observing a relationship as extreme as the observed relationship in the original data. I expect the test to produce a low simulated p-value, demonstrating the significance of the underlying relationship between age and OBP/SB. Following, I will utilize the bootstrap method to resample my data, producing a simulated confidence interval, providing insight into the relationship had the data been sampled differently. I expect the simulated confidence interval to include the observed relationship within the original data, demonstrating its significance. In parallel to my regression analysis, I will be assessing the significance of the relationship between age and OBP/SB. In contrast to the regression models, however, I will be assessing the relationship's significance using _simulated_ data.

## Data Collection and Cleaning

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from   scipy.optimize import curve_fit
from   sklearn.linear_model import LinearRegression
import statsmodels.api as sm
import scipy.stats as stats
import statistics
import random
```

```python
mlb=pd.read_csv("mlbdata5.csv")
hof=pd.read_csv("Hall_of_Fame.csv", usecols=["Name","Inducted As"])
#preserve an unmodified copy of the primary dataset
mlb_copy=mlb.copy()
```

mlbdata5.csv is my primary dataset of interest. It includes hitting statistics for each indiviual player season in the history of Major League Baseball. Hall_of_Fame.csv is a short dataset of minimal importance. It will only be utilized to identify the players within my primary dataset that were inducted into the Baseball Hall of Fame.

```python
#lowercase column names
mlb.columns=[col.lower() for col in mlb.columns]
hof.columns=[col.lower() for col in hof.columns]
hof.columns=[col.replace(" ","_") for col in hof.columns]
```

```python
#create a binary variable for those indcuted into the Hall of Fame as a player
hof["hof"]=hof["inducted_as"].apply(lambda x: 1 if x=="Player" else 0)
#merge: adds a column to the dataset indicating if the player was inducted into the Hall of Fame
mlb=pd.merge(mlb,hof[["name","hof"]], on='name', how='left')
#players not in the hall of fame dataset are not in the hall of fame: dummy=0
mlb["hof"]=mlb["hof"].fillna(0)
```

In the hof (Hall of Fame) dataframe, I created a binary variable to identify those indcuted into the Hall of Fame as _players_. "Player" is specified because the dataset includes others inducted into the Hall of Fame for other roles, such as managers. I merge my primary dataframe (mlb) with a subset of hof, effectively only adding one new column to mlb, a binary variable indicating if the player was inducted into the Hall of Fame. Players who do not receive values in the added column are assigned a 0, signifying that they were not inducted into the Hall of Fame as players.

```python
print ("The uncleaned dataset contains {} rows (observations) and {} columns (variables).".format(mlb.shape[0],mlb.shape[1]))
```

```python
#Subset data to only include seasons in which batter made 300 or more plate appearences
mlb=mlb[mlb["pa"]>300]
print("After subsetting the data, there are {} rows (observations).".format(len(mlb)))
```

The number of plate appearences a batter makes is the number of times a batter has the opportunity to hit. I subset the data in this manner because offensive seasons with fewer than 300 plate appearences are far too short to gain any real insight. For reference, a full-time player makes upwards of 500 plate appearences per season. 

```python
#Sum of NaNs in each column
np.sum(mlb.isna())
```

There are a large number of NaNs in multiple columns. My primary concern is the age variable, as it is my primary variable of interest. Luckily, it does not have many NaNs. Thus, I will likely drop those observations. 

In addition to age, I plan to utilize the sb (stolen base) variable. As I do not wish to drop 600 observations, I will replace those NaNs with zeroes. This, of course, will affect my analysis of the variable. I will thus make a note of this each time I utilize the variable. 

The remaining variables with NaNs, however, are not particularly significant statistics in assessing player performance. Thus, I will drop those variables entirely from the dataset.

```python
mlb["sb"]=mlb["sb"].fillna(0)
mlb=mlb.drop(columns=['babip', 'xwoba', "ibb","so","gdp","cs","rbi"])
```

```python
#this allows me to see the latest year in which there is a NaN in the age column
nan_sub=mlb[mlb.age.isna()]
nan_sub.loc[nan_sub.season.idxmax(),"season"] 
```

The last season in which a player's age went unrecorded was 1909. As I expected, the NaNs in the age column are only from the earlier years of the league's existence.

```python
#drop observations with NaNs in age column
mlb=mlb.dropna(subset=["age"])
```

```python
#any more NaNs?
np.sum(mlb.isna())
```

```python
mlb.dtypes
```

```python
#cast age column to int
mlb["age"] = mlb["age"].astype(int)
```

```python
plt.hist(mlb.age)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Number of Seasons by Age")
plt.tight_layout()
plt.show()
```

```python
mlb.age.value_counts(ascending=False)
```

As illustrated by both the histogram and series above, the most common age for players among seasons played is between 25 and 29.

```python
#drop the observations for seasons with ages 17-18 and 43-45, as there are too few to gain any meaningful insight
mlb=mlb[(mlb["age"]>=19) & (mlb["age"]<43)]
```

```python
#lets take another look at the age histogram after the drops
plt.hist(mlb.age)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Number of Seasons by Age")
plt.tight_layout()
plt.show()
```

```python
#Create a new column for the number of seasons a player played
mlb["career_length"]=[mlb["playerid"].value_counts()[p_id]for p_id in mlb["playerid"]]
#Create a new OPS column (On-Base Plus Slugging), a commonly used statistic to assess offensive performance
mlb["ops"]=mlb["obp"]+mlb["slg"]
#Create a new weighted column for OBP. OBP (On-Base-Percentage) is an average, so it must be weighted by PA (Plate Apperences)
mlb["weighted_obp"]=mlb["obp"]*mlb["pa"]
#Create a new column for the age of the player in their first season in MLB
mlb["age_first_season"]=[mlb.loc[mlb[mlb["playerid"]==p_id]["season"].idxmin()]["age"] for p_id in mlb["playerid"]]
#Create a new column for the age of the player in their last season in MLB
mlb["age_last_season"]=[mlb.loc[mlb[mlb["playerid"]==p_id]["season"].idxmax()]["age"] for p_id in mlb["playerid"]]
#Create a new categorical column. 1 for those under age 26. 2 for those between 26 and 30. 3 for those above 30.  
mlb["age_range"]=mlb["age"].apply(lambda x: 1 if x<26 else (3 if x>30 else 2))
#Create a new column for total career war across all individual seasons
mlb["career_war"]=[mlb[mlb["playerid"]==p_id]["war"].sum() for p_id in mlb["playerid"]]
#Create a new column for average performance per season
mlb["avg_career_war"]=mlb["career_war"]/mlb["career_length"]
```

```python
print ("The cleaned dataset has {} rows (observations) and {} columns (variables).".format(mlb.shape[0],mlb.shape[1]))
```

```python
#Save final curated dataset
mlb.to_csv("mlb.csv")
```

#### Multiple Regression
$  WAR=\beta_0+\beta_1Age+\beta_2Age^2
+\beta_3Avg Career War+\beta_4Age First Season+\beta_5Age Last Season+\beta_6CareerLength+\varepsilon$

```python
mlb["age_sqr"]=mlb["age"]**2
mult_var=["age", "age_sqr","avg_career_war","age_first_season","age_last_season","career_length"]
x = mlb[mult_var]
y = mlb.war
b = sm.add_constant(x)
ols_m = sm.OLS(y, b).fit()
print(ols_m.summary())
```

My primary variables of interest are age and age squared. The remaining variables are potential confounders. I included a squared term for age because I believed its underlying relationship with overall player performance is quadratic. In other words, I expect performance to improve at the outset of one's career, peak at a particular age, and decline thereafter. The results of the regression indicate that both the age and age squared terms are highly significant, each with a p-value of zero. Thus, we can be highly confident that their underlying relationship with WAR (overall performance) differs from zero. That is, we can be highly confident of the existence of a relationship between age and player performance. Moreover, the coefficient for the age term is positive, confirming my belief that young players become more valuable as they age. Conversely, the coefficient for the age squared term is negative, also validating my prediction that players face a decline in performance as they reach the latter parts of their careers. As age is not the sole predictor of player performance, I controlled for an additional four variables. The included confounding variables are as follows: average career war, or the average WAR the player attained per season played; the age at which the player played his first season; the age at which the player played his last season; career length, or the number of seasons a player played. In addition to a player's age, one's performance in a particular season is highly correlated to one's typical seasonal performance. In other words, an MVP caliber player is more likely to produce a high WAR than his less valuable counterparts. Thus, I included average performance as a confounder. Evidently, this, too, is highly significant, with a p-value of zero. The sole insignificant coefficient is career length. This is likely due to its high correlation with average career WAR, as those whose careers span a longer period of time are more likely to perform at a higher level. As a result of this correlation (computed below), there is the potential issue of multicollinearity. However, as it is unrelated to the age coefficients, it is not a particularly dire concern.

```python
print ("The correlation between average career performance and career length is {:.3f}.".format(mlb["avg_career_war"].corr(mlb["career_length"])))
```

#### Quadratic Model (among only Hall-Of-Famers) 
$  WAR=\beta_0+\beta_1Age+\beta_2Age^2+\varepsilon$

```python
# define custom function to fit
def quadratic(X, a, b, c):
    '''Second-order polynomial function'''
    return a*X**2 + b*X + c

# fit function parameters
quad_opt, quad_cov = curve_fit(quadratic, mlb_hof['age'], mlb_hof['war'])
age_sorted =mlb_hof['age'].sort_values()
# plot results
sns.scatterplot(x="age", y="war", data=mlb_hof, alpha=0.5)   # observed data
plt.plot(age_sorted, quadratic(age_sorted, *quad_opt), c='red', alpha=0.75) # predicted data
plt.xlabel('Age')
plt.ylabel('Wins Above Replacement')
plt.tight_layout()
plt.show()
print(f"Quadratic model r^2: {np.corrcoef(quadratic(mlb_hof['age'], *quad_opt), mlb_hof['war'])[0,1]**2:.3f}")
```

Excluding confounding variables results in a relatively low $r^2$, though I primarily included this model to provide a visual of the underlying quadratic relationship between age and overall performance. As expected, the peak occurs at roughly age 28.

#### Fixed Effects Regression Model (among only Hall-Of-Famers) 
#### $WAR$<sub>it</sub>-<overline> $\overline WAR $<sub>i</sub></overline>=$(\alpha_t- \overline\alpha)+\beta_1$($Age$<sub>it</sub>-$\overline Age$<sub>i</sub>)$+\beta_2$($Age$<sub>it</sub>$^2$-$\overline Age$<sub>i</sub>$^2$)$+(\varepsilon$<sub>it</sub>$- \overline \varepsilon$<sub>i</sub>)
i=individual (player), t=time (season)

```python
#individual means 
mlb["avg_career_age"]=[mlb[mlb["playerid"]==p_id]["age"].sum() for p_id in mlb["playerid"]]/mlb["career_length"]
mlb["avg_career_war"]=mlb["career_war"]/mlb["career_length"]

#subtract indiviual means from each observation 
mlb["demeaned_age"]=mlb["age"]-mlb["avg_career_age"]
mlb["demeaned_war"]=mlb["war"]-mlb["avg_career_war"]
```

```python
mlb_hof=mlb[mlb["hof"]==1]
# fit function parameters
quad_opt, quad_cov = curve_fit(quadratic, mlb_hof['demeaned_age'], mlb_hof['demeaned_war'])
age_sorted =mlb_hof['demeaned_age'].sort_values()
# plot results
sns.scatterplot(x="demeaned_age", y="demeaned_war", data=mlb_hof, alpha=0.5)   # observed data
plt.plot(age_sorted, quadratic(age_sorted, *quad_opt), c='red', alpha=0.75) # predicted data
plt.xlabel('Demeaned Age')
plt.ylabel('Demeaned Wins Above Replacement')
plt.tight_layout()
plt.show()
print(f"Quadratic model r^2: {np.corrcoef(quadratic(mlb_hof['demeaned_age'], *quad_opt), mlb_hof['demeaned_war'])[0,1]**2:.3f}")
```

An underlying assumption for regression models is the independence of observations. With the exception of the following regression model, however, a major limitation of my models is the dependence of the observations within the dataset. As the dataset includes multiple observations per subject, independence is not attainable. A player's performance in each season played is directly dependent on the player himself and his personal traits (ie, innate ability, durability). For example, in any particular season, better players are more likely to perform at a higher level and worse players are more likely to perform at a lower level. For this reason, I developed a fixed effects model. In doing so, I use the variation _within_ each subject to estimate age's effect on performance. To achieve this goal, I computed each player's average career WAR and average career age. Following, I subtracted these individual level means from each observation, producing a demeaned value for the variables of interest. In performing these computations, the model captures the individual fixed effect, or each player's core personality traits. The results are promising, as the coefficient of determination is almost doubled. Despite the high variability remaining within the dataset, the visual appears to support the presence of a quadratic relationship.

### Permutations and Bootstraps

```python
def permute(input_array):
    # shuffle is inplace, so copy to preserve input
    permuted = input_array.copy().values #convert to numpy array, avoiding warning
    np.random.shuffle(permuted)
    return pd.Series(permuted) # convert back to pandas
def plot_model_line(data, model, is_resampled=True):
    '''Takes a dataframe and a fitted model
    Plots a line of best fit to the data'''
    plt.plot(data['age'], data['age'] * model.coef_[0] + model.intercept_, color="steelblue", alpha=.7)
```

#### Wins Above Replacement (WAR)

##### Permutations

```python
# model true data
model = LinearRegression().fit(mlb[['age',"age_sqr"]], mlb["obp"])

# what fraction of permuted (null hyopthesis) slopes are greater than our observed slope?
steeper_slopes = 0
steeper_slopes2 = 0
n_permutations = 1000
permuted_slopes = np.zeros(n_permutations) # preallocate array of correct size
permuted_slopes2 = np.zeros(n_permutations) # preallocate array of correct size

for i in range(n_permutations):
    permuted_model = LinearRegression().fit(mlb[['age',"age_sqr"]], permute(mlb["obp"]))
    
    permuted_slopes[i] = permuted_model.coef_[0]
    
    if np.abs(permuted_model.coef_[0]) > np.abs(model.coef_[0]):
        steeper_slopes += 1
        
    permuted_slopes2[i] = permuted_model.coef_[1]
    
    if np.abs(permuted_model.coef_[1]) > np.abs(model.coef_[1]):
        steeper_slopes2 += 1
    
print(f'Age: Number of permuted slopes more extreme than observed slope: {steeper_slopes}')
print('Age: Simulated p (slope):', steeper_slopes / n_permutations)
print ("\n")
print(f'Age_sqr: Number of permuted slopes more extreme than observed slope: {steeper_slopes2}')
print('Age_sqr: Simulated p (slope):', steeper_slopes2 / n_permutations)
```

The regression models implemented above suggested that the nature of the relationship between age and WAR is quadratic. The above simulations confirm such findings. 1000 permutations were conducted, and virtually zero of the permuted slopes were as extreme or more extreme than the observed slope. Evidently, both the age and age squared terms are highly significant, upholding the results from the previous regressions, and solidifying the existence of a quadratic relationship between performance and age.

#### On Base Percentage (OBP)

##### Permutations

```python
# plot 100 iterations of models on permuted data
for i in range(100):
    permuted_model = LinearRegression().fit(mlb[["age"]], permute(mlb['obp']))
    plot_model_line(mlb, permuted_model)

# model true data and plot
model = LinearRegression().fit(mlb[["age"]], mlb["obp"])
plot_model_line(mlb, model, is_resampled=False)
plt.xlabel("Age")
plt.ylabel("On Base Percentage (OBP)")
plt.show()
```

```python
# what fraction of permuted (null hyopthesis) slopes are greater than our observed slope?
steeper_slopes = 0
n_permutations = 1000
permuted_slopes = np.zeros(n_permutations) # preallocate array of correct size

for i in range(n_permutations):
    permuted_model = LinearRegression().fit(mlb[['age']], permute(mlb["obp"]))
    
    permuted_slopes[i] = permuted_model.coef_[0]
    
    if np.abs(permuted_model.coef_[0]) > np.abs(model.coef_):
        steeper_slopes += 1
    
print(f'Number of permuted slopes more extreme than observed slope: {steeper_slopes}')
print('Simulated p (slope):', steeper_slopes / n_permutations)
```

```python
# plot where our observed slope falls in the distribution 
#  of slopes under the null hypothesis
binsizes, _, _ = plt.hist(permuted_slopes, bins=50)
plt.vlines(model.coef_[0], ymin=0, ymax=np.max(binsizes), color='red')
plt.show()
```

##### Boostraps

```python
n_bootstrap = 1000
bootstrap_slopes = np.zeros(n_bootstrap)
for i in range(n_bootstrap):
    resampled_data = mlb.sample(n=len(mlb), replace=True)
    resampled_model = LinearRegression().fit(resampled_data[["age"]], resampled_data["obp"])
    bootstrap_slopes[i] = resampled_model.coef_[0]
x=(np.quantile(bootstrap_slopes, q=[0.025, 0.975]))
print("The simulated 95% confidence interval is ({:.3f}, {:.3f}).".format(x[0],x[1]))
```

Previously, we observed a positive relationship between age and OBP, albeit an exceedingly small one. The above simulations confirm such findings. 1000 permutations were conducted, and zero of the permuted slopes were as extreme or more extreme than the observed slope. Additionally, the bootstraps produced a simulated confidence interval that does not include zero. Thus, we can be highly confident of the presence of a significant and positive underlying relationship. Evidently, plate discipline increases with age.

## Limitations

There are a number of limitations with my dataset, some significant and others less so. First, in the earlier years of Major League Baseball, there were a number of statistics that were not recorded. Comprehensive data from earlier years is lacking. Thus, the data is incomplete, and some columns have missing values. While I had dealt with these issues in my data cleaning, their effect on my analysis remains. The reason for this is that some of the complete columns, such as Wins Above Replacement (WAR), incorporate some of the missing values in their calculation. For example,  the formula for WAR contains a plethora of inputs, such as Stolen Bases (SB) and Runs Batted In (RBI). However, the SB and RBI columns contained a number of incomplete values. So although I handled this in my cleaning, its effect on WAR remains. Given WAR’s significance in assessing player performance, the incomplete values in other columns limits my analysis.

My dataset contains observations for each individual season among all players since the inception of the league. However, the data fails to include information as to the players’ whereabouts for seasons in which they do not play. Some may be sent down to the minors, others may retire, and others may have signed a deal to play professional baseball outside of the U.S. This, however, is likely the least significant limitation of my dataset, as these occurrences likely indicate that the player’s performance is no longer on par with the league. The one exception to this is injuries. A player may fail to participate due to an injury, and not because of his performance. This is not a particularly dire concern, however, as its prevalence is likely low.

As we observed in the exploratory analysis section, the dataset is not evenly distributed across ages. For example, there are significantly more age-27 and age-28 seasons than there are age-40 and age-41 seasons. Thus, computed averages for the latter two will be increasingly susceptible to outliers. Also, those still playing at later ages are comparatively different than those who stop playing at a younger age.

In the late 1990s and early 2000s, players across MLB tested positive for performance-enhancing drugs. As a result, a number of older players successfully sustained a high level of performance, sometimes superseding their performance in the prime of their careers. Given that there are already too few observations of those playing at age 35 and older, these outliers can possibly skew the data for that age range.  

My variable of interest, age, is the age the player is on July 1st of the given season. Thus, a player may be in his age-32 season, yet play upwards of half the season as a 31-year old. Thus, the age column is an imperfect representation of a player’s current age. While one might argue that a deviation of one year might be insignificant, given the limited range of ages within the dataset and the variable’s significance in my analysis, it is at least something to make note of.

```python
mlb = pd.concat([mlb, pd.get_dummies(mlb["age_range"], prefix="age_range")], axis='columns')

mult_var=["age_range_1","age_range_2","avg_career_war","age_first_season","age_last_season","career_length"]
x = mlb[mult_var]
y = mlb.war
b = sm.add_constant(x)
ols_m = sm.OLS(y, b).fit()
print(ols_m.summary())
```

In addition to the fixed effects model, I developed a model using dummy variables for age. The age_range_1 coefficient is a measure of performance for those below the age of 26 relative to those above the age of 30, and the age_range_2 coefficient is a measure of performance for those between the ages of 26 and 30 relative to those above 30.
